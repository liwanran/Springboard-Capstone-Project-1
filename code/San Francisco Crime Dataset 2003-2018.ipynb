## As the most fields in ":computed_region_" columns are null values, we remove these columns below by dropping them.

```python
# remove irrelevant columns
p1.drop(['PdId',
        ':@computed_region_yftq_j783', 
         ':@computed_region_p5aj_wyqh',
         ':@computed_region_rxqg_mtj9',
        ':@computed_region_bh8s_q3mv', 
        ':@computed_region_fyvs_ahh9',
        ':@computed_region_9dfj_4gjx',
        ':@computed_region_n4xg_c4py',
        ':@computed_region_4isq_27mq',
        ':@computed_region_fcz8_est8',
        ':@computed_region_pigm_ib2e',
        ':@computed_region_9jxd_iqea',
        ':@computed_region_6pnf_4xz7',
        ':@computed_region_6ezc_tdp2',
        ':@computed_region_h4ep_8xdi',
        ':@computed_region_nqbw_i6c3',
        ':@computed_region_2dwj_jsy4'], axis=1, inplace=True)
p1.head()
```

```python
# convert p1['date'] from string to datetime
p1['Date'] = pd.to_datetime(p1['Date'])
p1.info()
```

## Extract crime cases from from 2003 to 2017 from this dataset.

```python
p2003_2017 = p1[p1['Date'].dt.year < 2018]
p2003_2017.head()
```

```python
p2003_2017.shape
```

# 2. This dataset reflects the crime incidents that occurred in San Francisco from 1/1/2018 to 4/18/2019. Source: https://catalog.data.gov/dataset/police-department-incident-reports-2018-to-present

```python
# import dataset
p2 = pd.read_csv('datasets/major/Police_Department_Incident_Reports__2018_to_Present.csv')
p2.head()
```

```python
p2.info()
```

```python
# show all column names
p2.columns
```

```python
p2.shape
```

```python
p2.isnull().sum()
```

```python
# remove irrelavant columns
p2.drop(['Incident Datetime', 'Incident Year', 'Report Datetime', 'Filed Online', 'Row ID', 'Incident ID', 
         'CAD Number', 'Report Type Code', 'Report Type Description', 'Incident Code', 'CNN', 
         'HSOC Zones as of 2018-06-05', 'OWED Public Spaces', 'Incident Subcategory',
        'Central Market/Tenderloin Boundary Polygon - Updated', 'Supervisor District','Analysis Neighborhood',
        'Parks Alliance CPSI (27+TL sites)'], axis=1, inplace=True)
p2.head()
```

```python
p2.shape
```

```python
# change column names to be consistent with p1
p2.columns = ['Date', 'Time', 'DayOfWeek', 'IncidntNum', 'Category', 'Descript', 'Resolution',
              'Address', 'PdDistrict', 'Y', 'X', 'Location', 'SF Find Neighborhoods', 
              'Current Police Districts',  'Current Supervisor Districts', 'Analysis Neighborhoods']
p2.head()
```

```python
# convert p2['date'] from string to datetime
p2['Date'] = pd.to_datetime(p2['Date'])
p2.info()
```

```python
p_2018 = p2[p2['Date'].dt.year < 2019]
p_2018.head()
```

# 3. Combine the 2 datasets

```python
# Combine the 2 datasets and reset the indexes
p00 = pd.concat([p2003_2017, p_2018], ignore_index=True)
p00.head()
```

```python
p00.isnull().sum()
```

```python
p00.shape
```

# to remove duplicated crime incidents by using 'IncidntNum' and keeping the first row of the duplicated ones and keep the 1st row when there's a duplicate as the category was listed in the order of severity.

```python
p0 = p00.drop_duplicates('IncidntNum')
p0.head()
```

```python
p0.shape
```

```python
p0.isnull().sum()
```

## Save the new merged dataset as "p0".

```python
p0.to_csv('datasets/p0.csv')
```
